### Use beautifulsoup to pull the pokemon's name, number, and fighting type out of the elements in the table

```python
import urllib2
from bs4 import BeautifulSoup
import json
import re
import pandas as pd
soup = BeautifulSoup(urllib2.urlopen('http://www.serebii.net/diamondpearl/pokemon.shtml').read())
```

### Put the pokemon data into a dataframe and print it

```python
pokemon = {
    'number': [],
    'name': [],
    'type1': [],
    'type2': []
}

tables = soup.find_all('table', width='98%')
for id1, table in enumerate(tables):
    #print row
        for id2, row in enumerate(table.find_all('tr')):
            if id2 > 1:
                #print "<--- row --->", id2
                for id3, col in enumerate(row.find_all('td')):
                    if id3 == 0:
                        #print col.text.replace('\n', "").replace('\t',"").replace('\r',"")
                        pokemon['number'] += [col.text.replace('\n', "").replace('\t',"").replace('\r',"")]
                    if id3 == len(row.find_all('td'))-4:
                        #print col.text.split('\n')[1]
                        pokemon['name'] += [col.text.split('\n')[1]]
                    if id3 == len(row.find_all('td'))-2:
                        key = col.find('a').contents
                        pokemon['type1'] += [key[0]['src'][17:-4]]
                    if id3 == len(row.find_all('td'))-1:
                        key1 = col.find('a').contents
                        #print key1[0]['src']
                        pokemon['type2'] += [key1[0]['src'][17:-4]]
```
